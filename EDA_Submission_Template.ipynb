## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Handling missing values:
df['state'] = df['state'].fillna('Unknown')
df['work_interfere'] = df['work_interfere'].fillna('No response')
df['self_employed'] = df['self_employed'].fillna("Don't know")

df.isnull().sum()
````

````python
import numpy as np


def clean_gender(gender):
    gender = str(gender).strip().lower()

    if any(x in gender for x in ['female', 'f', 'woman']):
        return 'Female'
    elif any(x in gender for x in ['male', 'm', 'man']):
        return 'Male'
    elif any(x in gender for x in ['trans', 'genderqueer', 'gender fluid', 'non-binary', 'nonbinary', 'androgyne', 'agender', 'enby']):
        return 'Nonbinary'
    else:
        return 'Other'

df['Gender'] = df['Gender'].apply(clean_gender)


df.nunique()
````

````python
#converted work_interfere as categorical data:
interfere_order = ['No response', 'Never', 'Rarely', 'Sometimes', 'Often']
df['work_interfere'] = pd.Categorical(df['work_interfere'], categories=interfere_order, ordered=True)

#final look:
df
````

### What all manipulations have you done and insights you found?

####Handled missing values

1. Filled missing values in state with "Unknown" since state information is helpful but not critical.

2. Imputed missing work_interfere responses with "No response" to keep those respondents in analysis.

3. Filled missing self_employed entries with "Don't know" to avoid dropping rows unnecessarily.

4. Did not do any any changes in comment variable as it was moreover like a review column and is not necessarily required in the analysis

####Cleaned gender column

5. Standardized messy and inconsistent entries in Gender by writing a keyword-based function that mapped responses like "male-ish", "F", or "Trans man" into consistent categories: "Male", "Female", "Nonbinary", and "Other".

####Converted categorical ordering

6. Transformed work_interfere into an ordered categorical variable with the logical sequence: No response → Never → Rarely → Sometimes → Often.
